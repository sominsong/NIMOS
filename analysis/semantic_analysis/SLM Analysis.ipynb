```python
# EID의 path별 syscall path 딕셔너리 만들기
import pandas as pd

df = pd.DataFrame(columns = ['EID','PATH_NUM', 'SYSCALL_PATH'])

for EID in EIDList:
    totalPath = pathList[EID]["path"]
    for func, pathSet in totalPath.items():
        for path in pathSet:
                df = df.append(pd.DataFrame([[EID, func, path]], columns=['EID','PATH_NUM', 'SYSCALL_PATH']), ignore_index=True)
df.to_pickle("path_df.pkl")
print(df)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import DBSCAN
import numpy as np

# 문서를 명사 집합으로 보고 문서 리스트로 치환 (tfidfVectorizer 인풋 형태를 맞추기 위해)
text = [" ".join(syscall) for syscall in df['SYSCALL_PATH']]

########
N_GRAM_MAX = 3

tfidf_vectorizer = TfidfVectorizer(min_df = 5, ngram_range=(1,N_GRAM_MAX), sublinear_tf =True)
tfidf_vectorizer.fit(text)
vector = tfidf_vectorizer.transform(text).toarray()
# vector = np.array(vector) # Normalizer를 이용해 변환된 벡터
model = DBSCAN(eps=0.5,min_samples=10, metric = "cosine")
# 거리 계산 식으로는 Cosine distance를 이용
result = model.fit_predict(vector)
df["cluster_result"] = result
df.head()

cluster_result = dict()
for cluster_num in set(result):
    # -1,0은 노이즈 판별이 났거나 클러스터링이 안된 경우
    if(cluster_num == -1 or cluster_num == 0): 
        continue
    else:
        cluster_result[cluster_num] = set()
        print("cluster num : {}".format(cluster_num))
        temp_df = df[df['cluster_result'] == cluster_num] # cluster num 별로 조회
        for path in temp_df['EID']:
            print(path)
            cluster_result[cluster_num].add(path)
        print()
```

```python
from sklearn.preprocessing import normalize
from sklearn.feature_extraction.text import TfidfVectorizer

# 데이터 읽어오기
df = pd.read_pickle('path_df.pkl')

########
N_GRAM_MAX = 4

# 문서를 명사 집합으로 보고 문서 리스트로 치환 (tfidfVectorizer 인풋 형태를 맞추기 위해)
text = [" ".join(syscall) for syscall in df['SYSCALL_PATH']]

tfidf_vectorizer = TfidfVectorizer(min_df = 1, ngram_range=(1,N_GRAM_MAX), sublinear_tf =True)
tfidf_vectorizer.fit(text)
vector = tfidf_vectorizer.transform(text).toarray()
# vector = normalize(vector) # Normalizer를 이용해 변환된 벡터
vector.shape
```

```python
# K-means로 n개 군집으로 문서 군집화시키기
from sklearn.cluster import KMeans

##### 클러스터 개수 설정
cluster_num = 20

kmeans = KMeans(n_clusters=cluster_num,  init='k-means++')
# 비지도 학습이니 feature로만 학습시키고 예측
cluster_label = kmeans.fit_predict(vector)

# 군집화한 레이블값들을 document_df 에 추가하기
df['label'] = cluster_label
print(df.sort_values(by=['label']))
```

```python
# 그래프로 clustering 나타내기
# https://medium.com/kuzok/news-documents-clustering-using-python-latent-semantic-analysis-b95c7b68861c

# lsa 적용
from sklearn.utils.extmath import randomized_svd
# U: document-concept matrix, VT: term-concept matrix, sigma: concept-concept matrix, vetctor: document-term matrix obtained by tf-idf
# n-components: number of concepts/topics
U, Sigma, VT = randomized_svd(vector, n_components=cluster_num, n_iter=100, random_state=122)

# terms: 시스템콜 n-gram 집합
terms = tfidf_vectorizer.get_feature_names() 
for i, comp in enumerate(VT):
    terms_comp = zip(terms, comp)
    sorted_terms = sorted(terms_comp, key= lambda x:x[1], reverse=True)[:7]

# umap 사용하여 고차원 데이터 2차원으로 시각화
import umap.umap_ as umap
X_topics=U*Sigma
embedding = umap.UMAP(n_neighbors=20, min_dist=0.5).fit_transform(X_topics)

custom_palette = sns.color_palette("hls", cluster_num)
p = sns.scatterplot(x=embedding[:, 0], y=embedding[:, 1], hue=cluster_label, palette=custom_palette, alpha = 0.4, legend = False);
p.set_title(f"k-mean clustering (k: {cluster_num})")
```

```python
# 클러스터 간의 핵심 단어 추출
def get_cluster_details(cluster_model, cluster_data, feature_names,
                       cluster_num, top_n_features=10):
    cluster_details = {}
    # 각 클러스터 레이블별 feature들의 center값들 내림차순으로 정렬 후의 인덱스를 반환
    center_feature_idx = cluster_model.cluster_centers_.argsort()[:,::-1]
    
    # 개별 클러스터 레이블별로 
    for cluster_num in range(cluster_num):
        # 개별 클러스터별 정보를 담을 empty dict할당
        cluster_details[cluster_num] = {}
        cluster_details[cluster_num]['cluster'] = cluster_num
        
        # 각 feature별 center값들 정렬한 인덱스 중 상위 10개만 추출
        top_ftr_idx = center_feature_idx[cluster_num, :top_n_features]
        top_ftr = [feature_names[idx] for idx in top_ftr_idx]
        # top_ftr_idx를 활용해서 상위 10개 feature들의 center값들 반환
        # 반환하게 되면 array이기 떄문에 리스트로바꾸기
        top_ftr_val = cluster_model.cluster_centers_[cluster_num, top_ftr_idx].tolist()
        
        # cluster_details 딕셔너리에다가 개별 군집 정보 넣어주기
        cluster_details[cluster_num]['top_features'] = top_ftr
        cluster_details[cluster_num]['top_featrues_value'] = top_ftr_val
        # 해당 cluster_num으로 분류된 파일명(문서들) 넣어주기
        filenames = cluster_data[cluster_data['label']==cluster_num]['EID']
        # filenames가 df으로 반환되기 떄문에 값들만 출력해서 array->list로 변환
        filenames = filenames.values.tolist()
        cluster_details[cluster_num]['EID'] = filenames
    
    return cluster_details

def print_cluster_details(cluster_details):
    for cluster_num, cluster_detail in cluster_details.items():
        print(f"#####Cluster Num: {cluster_num}")
        print()
        print("상위 10개 feature단어들:\n", cluster_detail['top_features'])
        print()
        print(f"Cluster {cluster_num}으로 분류된 문서들:\n{cluster_detail['EID'][:5]}")
        print('-'*20)

feature_names = tfidf_vectorizer.get_feature_names()
cluster_details = get_cluster_details(cluster_model=kmeans,
                                     cluster_data=df,
                                     feature_names=feature_names,
                                     cluster_num=cluster_num,
                                     top_n_features=10)
print_cluster_details(cluster_details)
```

```python
# Path 별 시스템콜 분포 확인

def make_topictable_per_doc(ldamodel, corpus):
    topic_table = pd.DataFrame()

    # 몇 번째 Path인지를 의미하는 Path 번호와 해당 Path의 토픽 비중을 한 줄씩 꺼내온다.
    for i, topic_list in enumerate(ldamodel[corpus]):
        path = topic_list[0] if ldamodel.per_word_topics else topic_list            
        path = sorted(path, key=lambda x: (x[1]), reverse=True)
        # 각 Path에 대해서 비중이 높은 토픽순으로 토픽을 정렬한다.
        # EX) 정렬 전 0번 Path : (2번 토픽, 48.5%), (8번 토픽, 25%), (10번 토픽, 5%), (12번 토픽, 21.5%), 
        # Ex) 정렬 후 0번 Path : (2번 토픽, 48.5%), (8번 토픽, 25%), (12번 토픽, 21.5%), (10번 토픽, 5%)
        # 48 > 25 > 21 > 5 순으로 정렬이 된 것.

        # 모든 Path에 대해서 각각 아래를 수행
        for j, (topic_num, prop_topic) in enumerate(path): #  몇 번 토픽인지와 비중을 나눠서 저장한다.
            if j == 0:  # 정렬을 한 상태이므로 가장 앞에 있는 것이 가장 비중이 높은 토픽
                topic_table = topic_table.append(pd.Series([int(topic_num), round(prop_topic,4), topic_list]), ignore_index=True)
                # 가장 비중이 높은 토픽과, 가장 비중이 높은 토픽의 비중과, 전체 토픽의 비중을 저장한다.
            else:
                break
    return(topic_table)

topictable = make_topictable_per_doc(ldamodel, corpus)
topictable = topictable.reset_index() # Path 번호을 의미하는 열(column)로 사용하기 위해서 인덱스 열을 하나 더 만든다.
topictable.columns = ['Path 번호', '가장 비중이 높은 토픽', '가장 높은 토픽의 비중', '각 토픽의 비중']
topictable[:10]
```
