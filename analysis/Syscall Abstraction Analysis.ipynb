```python
# hard와 soft syspath 불러오기
import pandas as pd
hard_ngram = pd.read_pickle("hard_dedup_ngramdf.pkl")
soft_ngram = pd.read_pickle("soft_dedup_ngramdf.pkl")
hard_ngram.head()
```

### 익스플로잇에 쓰인 시스템콜 목록 excel로 저장

```python
total_syscall = list()
for index, row in soft_ngram.iterrows():
    total_syscall.extend(row.Ngram_syscall)
total_syscall = list(set(total_syscall))
len(total_syscall)
```

####  syscall number: syscall name 딕셔너리 만들기

```python
import subprocess

syscallTbl = dict()
cmd = f'cat /tmp/x86_64.syscall'
syscallx86 = subprocess.check_output(cmd, shell=True).decode().strip()
syscallx86List = syscallx86.split("\n")

for s in syscallx86List:
    if s and "__NR_" in s:
        sysname = s.split()[1].replace("__NR_","")
        sysnum = s.split()[2]
        syscallTbl[sysnum] = sysname
syscallTbl["436"] = "vm86old"
syscallTbl
```

```python
for syscall in total_syscall:
    print(syscallTbl[syscall])
```
